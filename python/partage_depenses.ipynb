```python

"""
fonction du type de l'application Tricount qui sert à déterminer,
au sein d'un groupe, ce que chacun doit rembourser à l'autre 
afin de répartir de manière équilibrée les dépenses totales du groupe.

Expérimentez l'application en exécutant la fonction partage_depenses_app()

"""

def partage_depenses_app():
    ### définition variables
    participants_num = 1
    participants_noms = []
    participants_montant = []
    repartition_depenses = zip(participants_noms, participants_montant)
    montant_total = 0       
    
    ### inputs nom de chacun des participanst et de leurs dépenses respectives
    name = input("Quel est le prénom du participant numéro " + str(participants_num) + " ?")
    participants_noms.append(name)
    
    while True:
        try:
            montant = float(input("Quel montant a dépensé " + name + " ?"))
            break
        except ValueError:
            print("Oops!  That was no valid number.  Try again...")
    participants_montant.append(montant)
    
    montant_total += float(montant)
    
    
    participant_new = input("Y a t-il des participants suplémentaires ?" + "\n" + "Répondez par o ou n :")
    while participant_new == "o" or participant_new == "O" or participant_new == "N" or participant_new == "n":
        break
    else:
        print("Répondez par oui ou non (\"o\" ou \"n\")")
        participant_new = input("Y a t-il des participants suplémentaires ?" + "\n" + "Répondez par o ou n :")
          
    
    while participant_new == "O" or participant_new == "o":
        participants_num += 1
        name = input("Quel est le prénom du participant numéro " + str(participants_num) + " ? ")
        participants_noms.append(name)
        
        while True:
            try:
                montant = float(input("Quel montant a dépensé " + name + " ?"))
                break
            except ValueError:
                print("Oops!  That was no valid number.  Try again...")
        participants_montant.append(montant)
        montant_total += float(montant)
        participant_new = input("Y a t-il des participants suplémentaires ?" + "\n" + "Répondez par oui ou non :")
        while participant_new == "o" or participant_new == "O" or participant_new == "N" or participant_new == "n":
            break
        else:
            print("Répondez par oui ou non (\"o\" ou \"n\")")
            participant_new = input("Y a t-il des participants suplémentaires ?" + "\n" + "Répondez par o ou n :")
    
    ### Résultat des inputs: résumé des dépenses et répartition entre chacun des paritcipants
    print("\n" + "Voici un résumé des dépenses :" + "\n")
    print("le montant total dépensé est de : " + str(round(montant_total, 2)) + " Euros") 
    print("La répartion des dépenses est comme suit : "+ "\n" + str(list(repartition_depenses)) + "\n")
    montant_moyen_1 = montant_total / len(participants_noms)
    montant_moyen = round(montant_moyen_1,2)
                                          
    i = 0
    while i < int(len(participants_montant)):
        if participants_montant[i] < montant_moyen:
            print(participants_noms[i] + " doit un montant de " + str(round(montant_moyen - participants_montant[i],2)))
        if participants_montant[i] > montant_moyen:
            print(participants_noms[i] + " doit se faire rembourser de " + str(round(participants_montant[i] - montant_moyen, 2)))
        i +=1
    
# creation liste imbriquee noms/montant
    repartition_depenses_liste = []
    j = 0
    while j < len(participants_montant):
        repartition_depenses_liste += [[participants_noms[j], participants_montant[j]]]
        j += 1
        
    
    #creations listes separees debiteurs/crediteurs
    
    debiteurs_liste = []
    crediteurs_liste = []
    neutre_liste = []
    
    b = 0
    while b < int(len(participants_montant)):
        if participants_montant[b] < montant_moyen:
                debiteurs_liste += [[participants_noms[b], round(montant_moyen - participants_montant[b], 2)]]
        if participants_montant[b] > montant_moyen:
                crediteurs_liste += [[participants_noms[b], round(participants_montant[b] - montant_moyen, 2)]]
        if participants_montant[b] == 0:
                neutre_liste += [[participants_noms[b], round(participants_montant[b] - montant_moyen, 2)]]
        b += 1
    
    debiteur_liste_tri = sorted(debiteurs_liste, key = lambda nombre1: nombre1[1], reverse = True)
    crediteur_liste_tri = sorted(crediteurs_liste, key = lambda nombre2: nombre2[1], reverse = True)
    neutre_liste_tri = sorted(neutre_liste, key = lambda nombre3: nombre3[1], reverse = True)
    
    #liste la plus courte:
    liste_courte = 0
    len_liste_debiteur = len(debiteur_liste_tri)
    len_liste_crediteur = len(crediteur_liste_tri)
    if len_liste_debiteur <= len_liste_crediteur:
        liste_courte = debiteur_liste_tri
    if len_liste_debiteur > len_liste_crediteur:
        liste_courte = crediteur_liste_tri
    
    #print("liste tri debiteur: ", debiteur_liste_tri)
    #print("liste tri créditeur: ", crediteur_liste_tri)
    #print("liste tri neutre: ", neutre_liste_tri)
                
    # calcul repartition
    print("\n" + "Voici le bilan de ce que doit chacun :")
    if liste_courte == crediteur_liste_tri:
        i = 0
        while i < len(crediteur_liste_tri):
            j = 0    
            while j < len(debiteur_liste_tri):
                
                if debiteur_liste_tri[j][1] > crediteur_liste_tri[i][1] and crediteur_liste_tri[i][1] != 0:
                    print("\n" + debiteur_liste_tri[j][0] + " doit " + str(round(crediteur_liste_tri[i][1], 2)) + " Euros à " + crediteur_liste_tri[i][0])
                    debiteur_liste_tri[j][1] = debiteur_liste_tri[j][1] - crediteur_liste_tri[i][1]
                    crediteur_liste_tri[i][1] = 0
                    
                    
                if debiteur_liste_tri[j][1] < crediteur_liste_tri[i][1] and debiteur_liste_tri[j][1] != 0:
                    print("\n" + debiteur_liste_tri[j][0] + " doit " + str(round(debiteur_liste_tri[j][1], 2)) + " Euros à " + crediteur_liste_tri[i][0])
                    crediteur_liste_tri[i][1] = crediteur_liste_tri[i][1] - debiteur_liste_tri[j][1]
                    debiteur_liste_tri[j][1] = 0
                    
                    
                if debiteur_liste_tri[j][1] == crediteur_liste_tri[i][1] and crediteur_liste_tri[i][1] != 0:
                    print("\n" + debiteur_liste_tri[j][0] + " doit " + str(round(debiteur_liste_tri[j][1], 2)) + " Euros à " + crediteur_liste_tri[i][0])
                    debiteur_liste_tri[j][1] = 0
                    crediteur_liste_tri[i][1] = 0
                    
                j += 1
            
            i += 1
        
    if liste_courte == debiteur_liste_tri: 
        j = 0
        while j < len(debiteur_liste_tri):
            i = 0    
            while i < len(crediteur_liste_tri):
                
                if debiteur_liste_tri[j][1] > crediteur_liste_tri[i][1] and crediteur_liste_tri[i][1] != 0:
                    print("\n" + debiteur_liste_tri[j][0] + " doit " + str(crediteur_liste_tri[i][1]) + " Euros à " + crediteur_liste_tri[i][0])
                    debiteur_liste_tri[j][1] = debiteur_liste_tri[j][1] - crediteur_liste_tri[i][1]
                    crediteur_liste_tri[i][1] = 0
                    
                    
                if debiteur_liste_tri[j][1] < crediteur_liste_tri[i][1] and debiteur_liste_tri[j][1] != 0:
                    print("\n" + debiteur_liste_tri[j][0] + " doit " + str(debiteur_liste_tri[j][1]) + " Euros à " + crediteur_liste_tri[i][0])
                    crediteur_liste_tri[i][1] = crediteur_liste_tri[i][1] - debiteur_liste_tri[j][1]
                    debiteur_liste_tri[j][1] = 0
                    
                    
                if debiteur_liste_tri[j][1] == crediteur_liste_tri[i][1] and crediteur_liste_tri[i][1] != 0:
                    print("\n" + debiteur_liste_tri[j][0] + " doit " + str(debiteur_liste_tri[j][1]) + " Euros à " + crediteur_liste_tri[i][0])
                    debiteur_liste_tri[j][1] = 0
                    crediteur_liste_tri[i][1] = 0
                    
                i += 1
            
            j += 1
    
        
    #print(debiteur_liste_tri)
    #print(crediteur_liste_tri)
        
    
                    
                    
                    
            
                    
                
                    
            
            
        
            
    
    
    
    

    
```
